## Q2. 포켓몬 타입별 집계


**MySQL은 사용하지 말고, Python과 Pandas를 활용하여 해당 집계를 수행합니다.**


- Type 기준 단일 groupby → count
- `ratio` : 전체 대비 비율, `"xx.xx%"` 문자열 (소수 2자리)
- `tier_label` : count 상위 3개 → `"Top3"`, 그 외 → `"Others"`
- 정렬 : count 내림차순
- 최종 컬럼 : `Type, count, ratio, tier_label`


```python
agg = (
    tmp_df.groupby('Type', as_index=False)
          .size()
          .rename(columns={'size': 'count'})
          .sort_values('count', ascending=False)
          .reset_index(drop=True)
)

# ratio: '%%' 문자열, 소수 2자리
agg['ratio'] = (agg['count'] / agg['count'].sum() * 100).map(lambda x: f'{x:.2f}%')

# tier_label: 상위 3개 Top3 / 나머지 Others
agg['tier_label'] = ['Top3' if i < 3 else 'Others' for i in range(len(agg))]

agg = agg[['Type', 'count', 'ratio', 'tier_label']]
agg
```

# Q4. 함수 선언 (삭제 전/후 비교)
- `prep_function(df, drop_col, drop_data)` → 단일 DataFrame 반환
- 컬럼 : `Type, before, after, pct_change`
- `before==0` 행은 `pct_change = np.nan`, 정렬 : before DESC
- 시각화는 별도 셀에서 grouped bar


```python
import numpy as np


def prep_function(df, drop_col, drop_data):
    before = (
        df.groupby(drop_col, as_index=False).size()
          .rename(columns={'size': 'before', drop_col: 'Type'})
    )
    kept = df[~df[drop_col].isin(drop_data)]
    after = (
        kept.groupby(drop_col, as_index=False).size()
            .rename(columns={'size': 'after', drop_col: 'Type'})
    )
    result = before.merge(after, on='Type', how='left')
    result['after'] = result['after'].fillna(0).astype(int)
    result['pct_change'] = np.where(
        result['before'] == 0, np.nan,
        np.round((result['after'] - result['before']) / result['before'] * 100, 1),
    )
    result = result.sort_values('before', ascending=False).reset_index(drop=True)
    return result


result_df = prep_function(df=tmp_df, drop_col='Type', drop_data=['GRASS', 'POISON'])
result_df
```

```python
import matplotlib.pyplot as plt

types  = result_df['Type'].tolist()
before = result_df['before'].to_numpy()
after  = result_df['after'].to_numpy()
x = np.arange(len(types))
width = 0.4

fig, ax = plt.subplots(figsize=(14, 6))
ax.bar(x - width/2, before, width, label='before')
ax.bar(x + width/2, after,  width, label='after')
ax.set_xticks(x)
ax.set_xticklabels(types, rotation=45, ha='right')
ax.set_ylabel('count')
ax.set_title('Before vs After by Type')
ax.legend()
plt.tight_layout()
plt.show()
```
