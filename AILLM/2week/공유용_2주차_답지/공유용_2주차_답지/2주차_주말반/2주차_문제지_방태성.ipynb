## Q9. 데이터 시각화 strip(10점)

```python
''' 참고자료 : 4-05 strip.ipynb '''


import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset('iris')  # 데이터 불러오기
display(iris.head())

# 시각화
plt.figure(figsize=(8, 5))
sns.boxplot(x='species', y='petal_length', data=iris, hue='species')

plt.title('Petal Length')
plt.xlabel('species')
plt.ylabel('petal_length')
plt.legend(title='species')
plt.tight_layout()
plt.show()


# x축에는 species를 설정합니다.
# y축에는 petal_length를 설정합니다.
# 색상은 species로 합니다.
# 제목은 Petal Length로 설정합니다.

```

## Q10. 데이터 시각화 Bubble(10점)

```python
''' 정답 코드를 이어서 작성해주세요 '''

''' 참고자료 : 4-06 scatter.ipynb '''

# 버블 차트 시각화
# 차트 설정
# 가로축 sepal_length
# 세로축 seqal_width
# 버블 크기 petal_length
# 색상은 species


import seaborn as sns
import plotly.express as px

# 데이터 불러오기
iris = sns.load_dataset('iris')
display(iris.head())

# 버블 차트 시각화
fig = px.scatter(
    iris,
    x='sepal_length',
    y='sepal_width',
    size='petal_length',
    color='species',
    title='Bubble Chart: Sepal vs Petal'
)

fig.show()

```
